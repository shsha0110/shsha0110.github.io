```python
#| label: causal-inference-full-code
#| echo: true
#| fig-cap: "        "

import numpy as np
import pandas as pd
import dowhy
from dowhy import CausalModel
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# 1.   (Data Generation)
# ---------------------------------------------------------
np.random.seed(42)
N = 1000

# (1)  :  (Age)
age = np.random.randint(20, 70, size=N)

# (2)  :   (Exercise, 0 or 1)
#     
prob_exercise = 1 / (1 + np.exp(-(age - 45) / 20))
exercise = np.random.binomial(1, prob_exercise)

# (3)  :  (Cholesterol)
#     (+2.0 * Age)
#     (True Effect = -10)
cholesterol = 2.0 * age - 10 * exercise + np.random.normal(150, 10, size=N)

df = pd.DataFrame({'Age': age, 'Exercise': exercise, 'Exercise_Prob': prob_exercise, 'Cholesterol': cholesterol})

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Exercise_Prob', 
    y='Cholesterol', 
    data=df, 
    hue='Age', 
    palette='viridis', 
    alpha=0.6
)
plt.title("Exercise vs Cholesterol (Colored by Age)")
plt.legend(title='Age')
plt.show()
# ---------------------------------------------------------
# 2.   (Naive Comparison)
# ---------------------------------------------------------
#        
mean_exercise = df[df['Exercise'] == 1]['Cholesterol'].mean()
mean_no_exercise = df[df['Exercise'] == 0]['Cholesterol'].mean()
naive_effect = mean_exercise - mean_no_exercise

print("\n" + "="*50)
print(f"1.    (Naive Estimate): {naive_effect:.4f}")
print("   ->      ( )")
print("="*50)

# ---------------------------------------------------------
# 3. DoWhy    (Causal Inference)
# ---------------------------------------------------------

#  1:   (Causal Graph )
model = CausalModel(
    data=df,
    treatment='Exercise',
    outcome='Cholesterol',
    common_causes=['Age'] #   
)

#  2:  (Identification) -      
identified_estimand = model.identify_effect()

#  3:  (Estimation) -    
estimate = model.estimate_effect(
    identified_estimand,
    method_name="backdoor.propensity_score_stratification",
    method_params={'num_strata': 5, 'clipping_threshold': 5}
)

print("\n" + "="*50)
print(f"2.    (Causal Estimate): {estimate.value:.4f}")
print(f"   ->   (-10.0)  ")
print("="*50)

# ---------------------------------------------------------
# 4.   (Refutation)
# ---------------------------------------------------------
print("\n[ ]    ...")

# (1)    (Placebo Treatment)
#      0  
refute_placebo = model.refute_estimate(
    identified_estimand,
    estimate,
    method_name="placebo_treatment_refuter",
    method_params={"placebo_type": "permute_treatment"}
)
print(f"\n- Placebo Test (Expected 0): {refute_placebo.new_effect:.4f}")

# (2)     (Random Common Cause)
#        (-10)  
refute_random = model.refute_estimate(
    identified_estimand,
    estimate,
    method_name="random_common_cause"
)
print(f"- Random Cause Test (Expected ~ -10): {refute_random.new_effect:.4f}")

# ---------------------------------------------------------
# 5.  (Visualization)
# ---------------------------------------------------------
plt.figure(figsize=(12, 5))

# ()   
plt.subplot(1, 2, 1)
sns.boxplot(x='Exercise', y='Cholesterol', data=df)
plt.title(f"Naive Comparison\n(Effect: {naive_effect:.2f})")

# ()     ()
plt.subplot(1, 2, 2)
sns.scatterplot(x='Age', y='Cholesterol', hue='Exercise', data=df, alpha=0.5)
#  O/X  
sns.regplot(x='Age', y='Cholesterol', data=df[df['Exercise']==0], scatter=False, label='No Exercise', color='blue')
sns.regplot(x='Age', y='Cholesterol', data=df[df['Exercise']==1], scatter=False, label='Exercise', color='orange')
plt.title(f"Controlled Comparison by Age\n(True Effect: -10.0)")
plt.legend()

plt.tight_layout()
plt.show()
```
