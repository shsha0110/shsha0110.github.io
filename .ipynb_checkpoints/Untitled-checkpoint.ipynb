```python
import numpy as np
import pandas as pd
import dowhy
from dowhy import CausalModel
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
```

```python
np.random.seed(42)
N = 1000

# 1. Confounder: 나이 (Age)
# 20세~70세
age = np.random.randint(20, 70, size=N)

# 2. Treatment: 운동량 (Exercise)
# 나이가 많을수록 운동을 많이 함 (Z -> X)
exercise = 0.5 * age + np.random.normal(10, 5, size=N)

# 3. Outcome: 콜레스테롤 수치 (Cholesterol)
# 나이가 많을수록 높음 (Z -> Y), 운동을 많이 할수록 낮아짐 (X -> Y)
# True Causal Effect: 운동 계수는 -1.5 (운동이 1단위 늘면 콜레스테롤 1.5 감소)
cholesterol = 2.0 * age - 1.5 * exercise + np.random.normal(150, 10, size=N)

# DataFrame 생성
df = pd.DataFrame({
    'Age': age,
    'Exercise': exercise,
    'Cholesterol': cholesterol
})

print(df.head())
```

```python
# 단순 상관계수 확인
correlation = df['Exercise'].corr(df['Cholesterol'])
print(f"Naive Correlation between Exercise and Cholesterol: {correlation:.4f}")

# 시각화 (Simpson's Paradox)
sns.regplot(x='Exercise', y='Cholesterol', data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("Naive Observation: More Exercise -> Higher Cholesterol?")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(
    x='Exercise', 
    y='Cholesterol', 
    data=df, 
    hue='Age', 
    palette='viridis', 
    alpha=0.6
)

plt.title("Exercise vs Cholesterol (Colored by Age)")
plt.legend(title='Age')
plt.show()
```

```python
# 인과 모델 정의
model = CausalModel(
    data=df,
    treatment='Exercise',
    outcome='Cholesterol',
    common_causes=['Age'] # 교란 변수 명시
)

# 모델 시각화
model.view_model()
```

```python
# 인과 효과 식별
identified_estimand = model.identify_effect()

print(identified_estimand)
```

```python
# ---------------------------------------------------------
# 방법 1: 선형 회귀 (Back-door Adjustment)
# ---------------------------------------------------------

estimate_reg = model.estimate_effect(
    identified_estimand,
    method_name="backdoor.linear_regression"
)

print(f"Causal Estimate (Linear Regression): {estimate_reg.value:.4f}")
```

```python
# ---------------------------------------------------------
# 방법 2: GPS (Generalized Propensity Score)
# ---------------------------------------------------------

X_gps = sm.add_constant(df['Age'])
gps_model = sm.OLS(df['Exercise'], X_gps).fit()

# 단계 2: 예측값과 잔차(Residuals) 계산
pred_exercise = gps_model.predict(X_gps)
residuals = df['Exercise'] - pred_exercise
# plt.hist(residuals)

# 단계 3: 잔차의 정규분포 확률 밀도(PDF) 계산
# 평균 0, 표준편차는 잔차의 표준편차를 따르는 정규분포 가정
density = stats.norm.pdf(residuals, loc=0, scale=residuals.std())

# 단계 4: 역확률 가중치(Weights) 계산
# 밀도가 낮을수록(흔치 않은 케이스일수록) 가중치를 높임
weights = 1 / density

# 시각화: 가중치 분포 확인 (너무 큰 가중치가 없는지 체크)
# plt.hist(weights, bins=30)
# plt.title("Distribution of Inverse Probability Weights (GPS)")
# plt.xlabel("Weight")
# plt.show()
```

```python
# 방법 2: GPS (Generalized Propensity Score)
X_outcome = sm.add_constant(df['Exercise'])
wls_model = sm.WLS(df['Cholesterol'], X_outcome, weights=weights).fit()

print("=== GPS 적용 결과 ===")
print(f"Estimated Causal Effect: {wls_model.params['Exercise']:.4f}")
print(f"True Effect: -1.5")
```

```python
# Placebo Treatment Refuter: 가짜 처리를 주었을 때 효과가 0이 나와야 함
refute_results = model.refute_estimate(
    identified_estimand,
    estimate_reg,
    method_name="placebo_treatment_refuter"
)

print(refute_results)
```
